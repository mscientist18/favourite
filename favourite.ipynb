```python
def bass_line_freq(track):
    sample_track = list(track)

    # c-value
    est_mean = np.mean(sample_track)

    # a-value
    est_std = 3 * np.std(sample_track) / (math.sqrt(2))

    bass_factor = int(round((est_std - est_mean) * 0.005))

    return bass_factor

def butter_lowpass(cutoff, fs, order=5):
    nyq = 0.5 * fs
    normal_cutoff = cutoff / nyq
    b, a = butter(order, normal_cutoff, btype='low', analog=False)
    return b, a

def butter_lowpass_filter(data, cutoff, fs, order=5):
    b, a = butter_lowpass(cutoff, fs, order=order)
    y = lfilter(b, a, data)
    return y

x=3600000
temp=data[x:x+1000000]
# temp=temp+sum(temp)/len(temp)
# temp=[(i-i/5 if i<2 else i) for i in temp]
temp = butter_lowpass_filter(temp,bass_line_freq(temp),fs)

Audio(temp,rate=fs)
```

```python
from pydub import AudioSegment
from os import listdir
import numpy as np
import math

# song_dir = "songs"
attenuate_db = 0
accentuate_db = 2


def bass_line_freq(track):
    sample_track = list(track)

    # c-value
    est_mean = np.mean(sample_track)

    # a-value
    est_std = 3 * np.std(sample_track) / (math.sqrt(2))

    bass_factor = int(round((est_std - est_mean) * 0.005))

    return bass_factor

# for filename in listdir(song_dir):
sample = AudioSegment.from_mp3("01 - Alone Again.mp3")
filtered = sample.low_pass_filter(bass_line_freq(sample.get_array_of_samples()))

combined = (sample - attenuate_db).overlay(filtered + accentuate_db)

combined.export("export.wav", format="wav")
from pydub.playback import play

# play(combined)

fs,data=read("export.wav")
data=data[:,0]
Audio(data[:1000000],rate=fs)
```
